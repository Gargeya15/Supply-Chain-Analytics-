1.First We Will Make and Copy And Delete unwanted Columns 

```python
data_clean=df.copy()
```

```python
drop_cols=[
    'Customer Password',
    'Product Description',
    'Product Image',
    'Order Zipcode'
]
```

```python

data_clean.drop(columns=drop_cols, inplace=True)
```

```python
data_clean.shape
```

2.Then We Will Change Date Types 

```python
data_clean['order date (DateOrders)'] = pd.to_datetime(
    data_clean['order date (DateOrders)']
)
```

```python
data_clean['shipping date (DateOrders)'] = pd.to_datetime(
    data_clean['shipping date (DateOrders)']
)
```

```python
data_clean.dtypes.tail()
```

3.Now We Will Create Business Columns 

```python
data_clean['Delivery_Delay']=(
    data_clean['Days for shipping (real)']
    - data_clean['Days for shipment (scheduled)']
)
```

```python
data_clean['Order_Year'] = (
    data_clean['order date (DateOrders)'].dt.year
)
```

```python
data_clean['Order_Month'] = (
    data_clean['order date (DateOrders)'].dt.month_name()
)
```

```python
data_clean['Order_Quarter'] = (
    data_clean['order date (DateOrders)'].dt.quarter
)
```

```python
data_clean[
    ['Delivery_Delay',
     'Order_Year',
     'Order_Month',
     'Order_Quarter']
].head()
```

5.Now We Will Create Dimension Tables and Fact Table For Power Bi Analysis 

(1).Product Dimension

```python
dim_product = data_clean[
[
'Product Card Id',
'Product Name',
'Category Name',
'Department Name',
'Product Price'
]
].drop_duplicates()
```

```python
dim_product.shape
```

(2).Customer Dimension

```python
dim_customer = data_clean[
[
'Customer Id',
'Customer Segment',
'Customer City',
'Customer State',
'Customer Country'
]
].drop_duplicates()
```

```python
dim_customer.shape
```

(3).Date Dimension

```python
dim_date = data_clean[
[
'order date (DateOrders)',
'Order_Year',
'Order_Month',
'Order_Quarter'
]
].drop_duplicates()
```

```python
dim_date.shape
```

(4).Fact Orders 

```python
fact_orders = data_clean[
[
'Order Id',
'Customer Id',
'Product Card Id',
'Sales',
'Benefit per order',
'Order Item Quantity',
'Shipping Mode',
'Delivery_Delay',
'Late_delivery_risk',
'order date (DateOrders)'
]
]
```

```python
fact_orders.shape
```

```python
# Export Product Dimension
dim_product.to_csv(
    'Dim_Product.csv',
    index=False,
    encoding='utf-8'
)

# Export Customer Dimension
dim_customer.to_csv(
    'Dim_Customer.csv',
    index=False,
    encoding='utf-8'
)

# Export Date Dimension
dim_date.to_csv(
    'Dim_Date.csv',
    index=False,
    encoding='utf-8'
)

# Export Fact Table
fact_orders.to_csv(
    'Fact_Orders.csv',
    index=False,
    encoding='utf-8'
)
```
